```
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
from scipy.interpolate import BSpline
from google.colab import files
```

```
x=0
f2=1-48*(x)+218*(x*x)-315*(x*x*x)+145*(x*x*x*x)
sigma2=-4*(x*x)+4*(x)+0.5
```

```
freedon=10
X=[]
F2=[]
Sigma2=[]
for i in range(0, freedon+1):
    x=i/freedon
    X.append(x)
    f2=1-48*(x)+218*(x*x)-315*(x*x*x)+145*(x*x*x*x)
    sigma2=-4*(x*x)+4*(x)+0.5
    F2.append(f2)
    Sigma2.append(sigma2)
cs10 = CubicSpline(X, F2)
xs = np.arange(0, 1, 0.01)
k = 4 #degree
Bspl10 = BSpline(X, F2, k)
plt.plot(X, F2, 'o', label='data')
plt.plot(xs, cs10(xs), label="NCS-10")
plt.plot(xs, Bspl10(xs), label="BS-10")
plt.legend(loc='upper right', ncol=1)
```
